```python
from ultralytics import YOLO
```

```python
# 1. Carregar a arquitetura pré-treinada (Transfer Learning)
# Usar um modelo 'nano' (-n) ou 'small' (-s) é ideal para datasets pequenos
model = YOLO('yolov8s-seg.pt') # '-seg' indica que é para Instance Segmentation
```

```python
print("Iniciando o treinamento do Modelo 2...")

results = model.train(
    data='BUSI_division/data.yaml', # O arquivo exportado do Roboflow
    
    # --- Controle de Épocas e Paciência ---
    epochs=150,              # Quantas vezes a IA vai ver o dataset inteiro
    patience=25,             # Early Stopping: se não melhorar por 25 épocas, ele para sozinho para não dar overfitting
    
    # --- Controle de Hardware e Lotes ---
    batch=16,                # Quantas imagens processar por vez (abaixe para 8 se faltar memória na placa de vídeo)
    imgsz=640,               # Redimensionamento padronizado das imagens
    
    # --- Controle de Aprendizado (O "Cérebro") ---
    lr0=0.001,               # Taxa de aprendizado inicial (o tamanho do "passo" que a IA dá ao aprender)
    optimizer='AdamW',       # Otimizador escolhido (AdamW é excelente para datasets menores)
    
    # --- Controle de Data Augmentation Direto no Código ---
    # Aqui você sobrescreve qualquer augmentation do Roboflow e aplica dinamicamente
    flipud=0.0,              # 0.0 = NUNCA espelhar de cabeça para baixo (Regra do ultrassom!)
    fliplr=0.5,              # 50% de chance de espelhar horizontalmente
    degrees=15.0,            # Rotação máxima de 15 graus
    hsv_s=0.1,               # Variação sutil na saturação
    hsv_v=0.1,               # Variação sutil no brilho/contraste
    
    # --- Validação ---
    val=True,                # Avaliar o modelo no conjunto de validação a cada época
    plots=True               # Gera gráficos automáticos de Loss, Precision e Recall na pasta final!
)

print("Treinamento finalizado. Cheque a pasta 'runs/segment/train' para ver os gráficos.")
```
